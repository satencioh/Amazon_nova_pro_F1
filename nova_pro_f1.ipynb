A continuacion tomamos el caso de uso de un video resumen de una carrera cualquiera de Formula 1 con el fin de conocer mas sobre este deporte mediante la explicacion y analisis que nos proporcionara Amazon Nova Pro.

Este ejercicio fue pensando para aquellos novatos en Formula 1 como yo üçë‚ô•Ô∏è

```python
import boto3
import json

# Create a Bedrock Runtime client in the AWS Region of your choice.
client = boto3.client(
    "bedrock-runtime",
    region_name="us-east-1",
)

MODEL_ID = "us.amazon.nova-pro-v1:0"


# Define user input
input_prompt = (
"""
¬°Hola! Soy nuevo en la F√≥rmula 1 y estoy viendo este video resumen de una carrera. La verdad, ¬°no entiendo casi nada de lo que pasa!

¬øPodr√≠as ser mi gu√≠a personal y enforcar tu explicacion netamente en este video? Imagina que es la primera vez que veo una carrera de F1. Presta mucha atenci√≥n a lo que sucede en el video y expl√≠came los siguientes puntos con un lenguaje muy sencillo y claro, sin asumir que s√© algo:

- ¬øQu√© autos se ven con frecuencia y qu√© acciones clave est√°n realizando en las tomas que aparecen?
- ¬øQu√© informaci√≥n clave est√°n transmitiendo los gr√°ficos, tablas o textos que aparecen en la pantalla de este video (por ejemplo, posiciones de los pilotos, diferencias de tiempo)?
- ¬øHay alg√∫n evento importante ocurriendo ahora mismo en el video (por ejemplo, adelantamientos, paradas en boxes, incidentes o el despliegue de banderas de color)? Explica qu√© significa cada uno y por qu√© es relevante dentro del contexto de esta carrera espec√≠fica si el video da pistas.
- ¬øQu√© partes del circuito se muestran en el video (curvas ic√≥nicas, la recta principal, la entrada o salida de boxes) y qu√© est√° ocurriendo all√≠ en este resumen?

Quiero que tu explicaci√≥n se centre en ayudarme a interpretar lo que estoy viendo en este video en particular.
"""
)


#S3 Location: Use for larger files, (Recommended for payload greater than 25MB upto 1GB)
s3_video_path = "s3://gpformula1/GrandPrix_f1.mp4"

# Video format, supported formats atm: MP4, MOV, MKV, WebM, FLV, MPEG, MPG, WMV, 3GP
vid_format = "mp4"

# Define a "user" message including both the image and a text prompt.
message_list = [
    {
        "role": "user",
        "content": [
            {
                "video": {
                    "format": vid_format,
                    "source": {
                        "s3Location": {
                            "uri": s3_video_path
                        }
                    }
                }
            },
            {
                "text": input_prompt
            }
        ]
    }
]

# Inference parameters
inf_params = {"maxTokens": 1000, "topP": 0.8, "temperature": 0.3}

try:
    response = client.converse(
        modelId=MODEL_ID,
        messages=message_list,
        inferenceConfig=inf_params
    )

    response_text = response["output"]["message"]["content"][0]["text"]
    print(response_text)

except Exception as e:
    print("Error al invocar el modelo:", str(e))
```
